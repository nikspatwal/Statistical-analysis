### Creating a new DataFrame of "Case-Death Ratio of Diseases" 

```python
x = pd.DataFrame(columns=['M','J','AD','AR','V'])
x['M'] = frame['Malaria Cases']/frame['Malaria Deaths']
x['J'] = frame['Japanese Encephalitis Cases']/frame['Japanese Encephalitis Deaths']
x['AD'] = frame['Acute Diarrhoeal Diseases*  Cases']/frame['Acute Diarrhoeal Diseases*  Deaths']
x['AR'] = frame['Acute Respiaratory *\nInfection Cases']/frame['Acute Respiaratory *\nInfection Deaths']
x['V'] = frame['Viral Hepatitis* Cases']/frame['Viral Hepatitis* Deaths']

_ = x.fillna(0, inplace=True)
u = x.drop(['M'])
print(type(u))
```

### Normalizing the new generated dataframe to "df_normal"

```python
hh = u.dtypes.index
hh
float_array = u.values.astype(float)
min_max_scaler = preprocessing.MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(float_array)
df_normal = pd.DataFrame(scaled_array)
df_normal
```

### Showing Bar Graph for the new Normalized Death-Cases Ratio for 15 years

```python
ux = df_normal[range(0,5)].plot(kind='bar', title ="CASES-DEATH RATIO", figsize=(15, 10), legend=True, fontsize=12)
ux.set_xlabel("Years", fontsize=12)
ux.set_ylabel("NUMBER OF CASES and DEATHS", fontsize=12)
plt.legend()
plt.show()
```
